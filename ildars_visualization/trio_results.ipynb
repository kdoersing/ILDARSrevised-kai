## CUMULATIVE PLOTTING

```python
# plotting x1-x2, y1-y2, z1-z2

import matplotlib.pyplot as plt
import numpy as np

column1 = np.asarray(df['computed position'])
column2 = np.asarray(df['original position'])
column3 = np.asarray(df['offset'])

fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection ='3d')
point_x = []
point_y = []
point_z = []
print(type(point_x))
# syntax for 3-D projection

for i in range(0,10):
   
    print('computed position', column1[i])
    

    split_row3 = column1[i].split()
    print(split_row3[0])
    print(split_row3[1])
    print(split_row3[2])
    
    print('original position', column2[i])
    split_row3_2 = column2[i].split()
    print(split_row3_2[0])
    print(split_row3_2[1])
    print(split_row3_2[2])
    
    print('offset',column3[i])

    # list of lists to be looped over
    point_x = ([float(split_row3[0]),float(split_row3_2[0])]) # x1 , x2
    point_y = ([float(split_row3[1]),float(split_row3_2[1])]) # y1, y2 
    point_z = ([float(split_row3[2]),float(split_row3_2[2])]) # z1, z2
    print('point x is',point_x)
    print('point y is',point_y)
    print('point z is',point_z)
    print(type(point_x),type(point_y),type(point_z))
    # Ensure that the next plot doesn't overwrite the first plot

    ax.plot3D(point_x[-1], point_y[-1], point_z[-1], 'ob') # original position
    ax.plot3D(point_x[0], point_y[0], point_z[0], 'og') # computed position
    ax.plot3D(point_x, point_y, point_z, 'red') # lines
    
    ax.set_title('point_display')
plt.show()
    

```
